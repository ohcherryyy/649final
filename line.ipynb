```python
import pandas as pd
import altair as alt
import numpy as np
import pprint
```

```python
df = pd.read_csv("vgsales.csv")
```

```python
df
```

```python
Nintendo_df = df[df.Publisher.str.contains('Nintendo') == True]
```

```python
Nintendo_df
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
df.groupby(['Genre'])['NA_Sales'].sum().sort_values(ascending=False)
```

```python
orderr = df.groupby(['Genre'])['NA_Sales'].sum().sort_values(ascending=False).to_frame().index.tolist()

line = alt.Chart(Nintendo_df, title="NA_Sales Trends by Genre").transform_aggregate(
    NA_Sales='sum(NA_Sales)',
    groupby=['Year','Genre']
).mark_line().encode(
    x = alt.X('Year:N'),
    y = alt.Y('NA_Sales:Q', title='NA_Sales (million $)'),
    color = alt.Color('Genre:N', title='Genre', sort=orderr)
)

chart = alt.layer(line).configure_title(fontSize=18, anchor="middle").configure_legend(titleFontSize=14,labelFontSize=12).configure_axis(titleFontSize=14,labelFontSize=11) 

NA_line = chart.encode(
    tooltip=['Year:N','Genre:N','NA_Sales:Q']
)

NA_line
```

```python
alt.data_transformers.enable('default', max_rows=None)
orderr = df.groupby(['Genre'])['NA_Sales'].sum().sort_values(ascending=False).to_frame().index.tolist()
line = alt.Chart(df, title="NA_Sales Trends by Genre").transform_aggregate(
    NA_Sales='sum(NA_Sales)',
    groupby=['Year','Genre']
).mark_line().encode(
    x = alt.X('Year:Q',scale=alt.Scale(domain=[1979,2019])),
    y = alt.Y('NA_Sales:Q', title='NA_Sales (million $)'),
    color = alt.Color('Genre:N', title='Genre', sort=orderr)
).transform_filter(alt.datum.Genre!="nan")

NA_line = line.encode(
    tooltip=['Year:N','Genre:N','NA_Sales:Q']
)

selectGenre=alt.selection_single(
    fields=['Genre'],
    init={"Genre":orderr[0]},
    bind=alt.binding_select(options=orderr,name="Select Genres:  ")
)

colorCondition = alt.condition(selectGenre,alt.Color("Genre:N"),alt.value('lightgray'))

NA_final = NA_line.add_selection(
    selectGenre
).encode(
    color=colorCondition
)

NA_final

```
