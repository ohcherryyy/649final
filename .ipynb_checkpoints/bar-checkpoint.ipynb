```python
import pandas as pd
import altair as alt
import numpy as np
import pprint
```

```python
df = pd.read_csv("vgsales.csv")
```

```python
Nintendo_df = df[df.Publisher.str.contains('Nintendo') == True]
```

```python
Nintendo_df
```

```python
rank50 =df.sort_values(by='Global_Sales', ascending=False).head(50)
rank50
```

```python
region_order = ['NA_Sales', 'JP_Sales', 'PAL_Sales', 'Other_Sales']
re = ['America', 'Japan', 'PAL', 'Other Regions']

bar = alt.Chart(rank50, title="Regional Sales of Top 50 Video Games",width=800).transform_fold(
  ['NA_Sales', 'JP_Sales', 'PAL_Sales', 'Other_Sales'],
  as_=['Regional_Sales', 'Amount']
).mark_bar().encode(
  x=alt.X('Publisher:N', sort='-y'),
  y=alt.Y('Amount:Q', title='Sales (million $)'),
  color=alt.Color('sum(Regional_Sales):N', sort=region_order),
  order=alt.Order('region_order:Q'),
    opacity=alt.condition(alt.datum.Publisher=='Nintendo  ',alt.value(1),alt.value(0.4))
)

chart = alt.layer(bar).configure_title(fontSize=18, anchor="middle").configure_legend(titleFontSize=14,labelFontSize=12).configure_axis(titleFontSize=14,labelFontSize=11) 

barr = chart.encode(
    tooltip=['Name:N','Regional_Sales:N','Amount:Q','Global_Sales:Q','Genre:N']
)

barr
```

```python
region_order = ['NA_Sales', 'JP_Sales', 'PAL_Sales', 'Other_Sales']
re = ['America', 'Japan', 'PAL', 'Other Regions']

point = alt.Chart(rank50, title="Regional_Sales of Top 50 Nintendo Video Games").transform_fold(
  ['NA_Sales', 'JP_Sales', 'PAL_Sales', 'Other_Sales'],
  as_=['Regional_Sales', 'Amount']
).mark_point().encode(
  x=alt.X('Name:N', sort='-y'),
  y=alt.Y('Amount:Q', title='Sales (million $)'),
  color=alt.Color('Regional_Sales:N', sort=region_order),
  order=alt.Order('region_order:Q')
)

chart = alt.layer(point).configure_title(fontSize=18, anchor="middle").configure_legend(titleFontSize=14,labelFontSize=12).configure_axis(titleFontSize=14,labelFontSize=11) 

pointt = chart.encode(
    tooltip=['Name:N','Regional_Sales:N','Amount:Q','Global_Sales:Q']
)

pointt
```
